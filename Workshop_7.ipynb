# Data Introduction
This dataset contains medical diagnostic information of females at least 21 years old of Pima Indian heritage. It is used to predict the likelihood of diabetes in patients based on various health attributes. It is useful for building predictive models, statistical analysis, and understanding key health indicators associated with diabetes risk.

Dataset Attributes:


1. Pregnancies: To express the Number of pregnancies
2. Glucose: To express the Glucose level in blood
3. BloodPressure: To express the Blood pressure measurement
4. SkinThickness: To express the thickness of the skin
5. Insulin: To express the Insulin level in blood
6. BMI: To express the Body mass index
7. DiabetesPedigreeFunction: To express the Diabetes percentage
8. Age: To express the age
9. Outcome: To express the final result 1 is YES o is NO

# EDA

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Count the occurrences of each outcome (0 or 1)
outcome_counts = df['Outcome'].value_counts()

# Create a bar chart using Seaborn
plt.figure(figsize=(8, 6))
sns.countplot(x='Outcome', data=df)
plt.title('Distribution of Diabetes Outcome')
plt.xlabel('Outcome (0 = No Diabetes, 1 = Diabetes)')
plt.ylabel('Number of Patients')
plt.show()
```

```python

# List of columns for bar graphs
categorical_cols = ['Pregnancies', 'Outcome']


# Function to create bar plots for each column

def plot_bar_graphs(df, col):
  plt.figure(figsize=(8, 6))
  sns.countplot(x=col, data=df)
  plt.title(f'Bar Plot of {col}')
  plt.xlabel(col)
  plt.ylabel('Count')
  plt.show()

# Create bar plots for specified columns
for col in categorical_cols:
  plot_bar_graphs(df, col)
```

```python
# Histograms for each numerical feature
plt.figure(figsize=(15, 10))
for i, col in enumerate(df.select_dtypes(include=['number']).columns):
    plt.subplot(3, 3, i + 1)
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')
plt.tight_layout()
plt.show()
```

```python
# Scatter plot for all numerical features against 'Outcome'
plt.figure(figsize=(15, 10))
for i, col in enumerate(df.select_dtypes(include=['number']).columns):
    plt.subplot(3, 3, i + 1)
    sns.scatterplot(x=col, y='Outcome', data=df)
    plt.title(f'Scatter plot of {col} vs Outcome')
plt.tight_layout()
plt.show()
```

```python
# Box plots for each numerical feature
plt.figure(figsize=(15, 10))
for i, col in enumerate(df.select_dtypes(include=['number']).columns):
    plt.subplot(3, 3, i + 1)
    sns.boxplot(y=df[col])
    plt.title(f'Box Plot of {col}')
plt.tight_layout()
plt.show()
```

```python
# Line chart for Glucose levels over pregnancies
plt.figure(figsize=(10, 6))
sns.lineplot(x='Pregnancies', y='Glucose', data=df)
plt.title('Glucose Levels over Pregnancies')
plt.xlabel('Number of Pregnancies')
plt.ylabel('Glucose Level')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix,mean_squared_error, r2_score
```

```python
# Data preprocessing - StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
# Data splitting
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.2, random_state=42)
```

```python
# Model training
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
# Predictions
y_pred = model.predict(x_test)
```

```python
# Evaluating the model
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse}")

r2 = r2_score(y_test, y_pred)
print(f"R-squared: {r2}")
```
