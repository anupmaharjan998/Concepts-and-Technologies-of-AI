# Pre Processing

```python
# Select relevant features and target
features = ["Order_Quantity", "Unit_Cost", "Unit_Price", "Profit", "Cost"]
target = "Revenue"

# Shuffle the data
shuffled_df = df.sample(frac=1, random_state=42).reset_index(drop=True)

# Calculate the split index (80% for training, 20% for testing)
split_index = int(0.8 * len(shuffled_df))

# Split the data into training and testing sets
train_data = shuffled_df[:split_index]
test_data = shuffled_df[split_index:]

# Separate features and target for training and testing
X_train = train_data[features]
y_train = train_data[target]

X_test = test_data[features]
y_test = test_data[target]
```

# Linear Regression Model

```python
# Linear Regression from scratch
class LinearRegressionScratch:
    def __init__(self):
        self.weights = None

    def fit(self, X, y):
        X = np.c_[np.ones(X.shape[0]), X]  # Add intercept
        self.weights = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)

    def predict(self, X):
        X = np.c_[np.ones(X.shape[0]), X]
        return X.dot(self.weights)

    def get_params(self, deep=True):
        return {}  # No parameters to tune for this simple model

    def set_params(self, **params):
        return self  # Nothing to set for now

```

```python
def mean_absolute_error_manual(y_true, y_pred):
    return np.mean(np.abs(y_true - y_pred))

def mean_squared_error_manual(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)

def r2_score_manual(y_true, y_pred):
    total_variance = np.sum((y_true - np.mean(y_true)) ** 2)
    residual_variance = np.sum((y_true - y_pred) ** 2)
    return 1 - (residual_variance / total_variance)
```

```python
# Train & evaluate custom Linear Regression model
model_scratch = LinearRegressionScratch()
model_scratch.fit(X_train, y_train)
y_pred_scratch = model_scratch.predict(X_test)

print("Custom Linear Regression Metrics:")
print("MAE:", mean_absolute_error_manual(y_test, y_pred_scratch))
print("MSE:", mean_squared_error_manual(y_test, y_pred_scratch))
print("R2 Score:", r2_score_manual(y_test, y_pred_scratch))
```

# Gradient Descent

```python
import numpy as np

class LinearRegressionGradientDescent:
    def __init__(self, learning_rate=0.001, iterations=1000):
        self.learning_rate = learning_rate
        self.iterations = iterations
        self.weights = None

    def fit(self, X, y):
        # Feature scaling (standardization) without sklearn
        X = self.feature_scaling(X)

        # Add intercept term (bias)
        X = np.c_[np.ones(X.shape[0]), X]

        # Initialize weights to zeros
        self.weights = np.zeros(X.shape[1])

        # Gradient Descent
        for i in range(self.iterations):
            # Prediction
            predictions = X.dot(self.weights)

            # Compute the error (residuals)
            error = predictions - y

            # Compute the gradient (derivative of MSE w.r.t weights)
            gradient = (2 / len(X)) * X.T.dot(error)

            # Update weights
            self.weights -= self.learning_rate * gradient

            # Optionally, you can check for NaN or very large updates:
            if np.any(np.isnan(self.weights)):
                print(f"NaN encountered in weights at iteration {i}")
                break

    def predict(self, X):
        # Feature scaling (standardization) without sklearn
        X = self.feature_scaling(X)

        # Add intercept term (bias)
        X = np.c_[np.ones(X.shape[0]), X]
        return X.dot(self.weights)

    def feature_scaling(self, X):
        """Manually scale features to zero mean and unit variance"""
        mean = np.mean(X, axis=0)
        std = np.std(X, axis=0)
        return (X - mean) / std

    # Implement get_params to work with GridSearchCV
    def get_params(self, deep=True):
        return {"learning_rate": self.learning_rate, "iterations": self.iterations}


     # Implement set_params to work with GridSearchCV
    def set_params(self, learning_rate=None, iterations=None):
        if learning_rate is not None:
            self.learning_rate = learning_rate
        if iterations is not None:
            self.iterations = iterations
        return self

    @staticmethod
    def mean_absolute_error_manual(y_true, y_pred):
        return np.mean(np.abs(y_true - y_pred))

    @staticmethod
    def mean_squared_error_manual(y_true, y_pred):
        return np.mean((y_true - y_pred) ** 2)

    @staticmethod
    def r2_score_manual(y_true, y_pred):
        total_variance = np.sum((y_true - np.mean(y_true)) ** 2)
        residual_variance = np.sum((y_true - y_pred) ** 2)
        return 1 - (residual_variance / total_variance)
```

```python
# Train the model
model_gd = LinearRegressionGradientDescent(learning_rate=0.001, iterations=1000)
model_gd.fit(X_train, y_train)

# Make predictions
y_pred_gd = model_gd.predict(X_test)

# Calculate metrics
mae = model_gd.mean_absolute_error_manual(y_test, y_pred_gd)
mse = model_gd.mean_squared_error_manual(y_test, y_pred_gd)
r2 = model_gd.r2_score_manual(y_test, y_pred_gd)

print("Gradient Descent Model Metrics:")
print("MAE:", mae)
print("MSE:", mse)
print("R2 Score:", r2)
```

# Hyperparameter Optimization and Cross-Validation Using sklearn

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error
from sklearn.base import BaseEstimator, RegressorMixin
```

```python
# Hyperparameter grid for Linear Regression
param_grid_lr = {
    'fit_intercept': [True, False],  # Try intercept or no intercept
}

# Initialize GridSearchCV
grid_search_lr = GridSearchCV(LinearRegressionScratch(), param_grid_lr, cv=5, scoring='neg_mean_squared_error')
grid_search_lr.fit(X_train, y_train)

# Get best model and hyperparameters
best_model_lr = grid_search_lr.best_estimator_
best_params_lr = grid_search_lr.best_params_

# Predict and evaluate on test data
y_pred_lr = best_model_lr.predict(X_test)
mse_lr = mean_squared_error(y_test, y_pred_lr)

# Results
print("Best Linear Regression Model Hyperparameters:", best_params_lr)
print("MSE on test data:", mse_lr)
```

```python
# Hyperparameter grid for Gradient Descent
param_grid_gd = {
    'learning_rate': [0.001, 0.01, 0.1],  # Test different learning rates
    'iterations': [500, 1000, 1500]  # Test different number of iterations
}

# Initialize GridSearchCV
grid_search_gd = GridSearchCV(LinearRegressionGradientDescent(), param_grid_gd, cv=5, scoring='neg_mean_squared_error')
grid_search_gd.fit(X_train, y_train)

# Get best model and hyperparameters
best_model_gd = grid_search_gd.best_estimator_
best_params_gd = grid_search_gd.best_params_

# Predict and evaluate on test data
y_pred_gd = best_model_gd.predict(X_test)
mse_gd = mean_squared_error(y_test, y_pred_gd)

# Results
print("Best Gradient Descent Model Hyperparameters:", best_params_gd)
print("MSE on test data:", mse_gd)
```
