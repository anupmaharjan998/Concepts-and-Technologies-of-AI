```python
#3.2 Exercises - Data Cleaning and Transformations with ”Titanic Dataset”:

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Titanic dataset
try:
    titanic_df = pd.read_csv("/content/drive/MyDrive/AI/Datasets/Titanic-Dataset.csv")
    print("Dataset loaded successfully.")
except FileNotFoundError:
    print("Error: 'titanic.csv' not found. Please check the file path.")
    exit()

```

```python
# Problem 1
subset_df = titanic_df[['Name', 'Pclass', 'Sex', 'Age', 'Fare', 'Survived']].copy()
subset_df = subset_df[subset_df['Pclass'] == 1]

fare_stats = subset_df['Fare'].agg(['mean', 'median', 'max', 'min'])
print("Fare Statistics for First-Class Passengers:\n", fare_stats)
```

```python
# Problem 2
null_age_count = subset_df['Age'].isnull().sum()
print("\nNumber of null values in 'Age' column:", null_age_count)
subset_df.dropna(subset=['Age'], inplace=True)
```

```python

# Problem 3
# Assuming 'Embarked' column is present in the titanic dataset and has values 'C', 'Q', or 'S'.

if 'Embarked' in titanic_df.columns:  # Check if the column exists
    embarked_dummies = pd.get_dummies(titanic_df['Embarked'], prefix='Embarked')
    titanic_df = pd.concat([titanic_df, embarked_dummies], axis=1)
    titanic_df.drop('Embarked', axis=1, inplace=True)
    print("\nFirst few rows of the modified DataFrame:\n", titanic_df.head())

else:
    print("\n'Embarked' column not found in the DataFrame.")
```

```python

# Problem 4
# Assuming 'Survived' and 'Sex' columns exist
if 'Survived' in titanic_df.columns and 'Sex' in titanic_df.columns:
    survival_by_sex = titanic_df.groupby('Sex')['Survived'].mean()
    print("\nMean survival rate by sex:\n", survival_by_sex)

    plt.figure(figsize=(8, 6))
    sns.countplot(x='Sex', hue='Survived', data=titanic_df)
    plt.title('Survival Distribution by Gender')
    plt.show()
else:
  print("Required columns for survival analysis not found.")
```

```python
#Problem 5 - Needs 'Embarked' columns from Problem 3 or original dataset.
if 'Embarked_C' in titanic_df.columns and 'Embarked_Q' in titanic_df.columns and 'Embarked_S' in titanic_df.columns and 'Survived' in titanic_df.columns:
  plt.figure(figsize=(10, 6))
  sns.countplot(x='Sex', hue='Survived', data=titanic_df, col='Embarked_C')
  plt.show()
else:
  print("'Embarked' columns or 'Survived' not found, cannot create visualization.")
```

```python
# Problem 6 (Optional) - Requires age quantiles
if 'Age' in titanic_df.columns and 'Pclass' in titanic_df.columns and 'Survived' in titanic_df.columns:
  titanic_df['AgeGroup'] = pd.qcut(titanic_df['Age'], 5, labels=False) #create age groups

  survival_by_age_class = titanic_df.groupby(['AgeGroup', 'Pclass'])['Survived'].mean().unstack()
  print(survival_by_age_class)

  survival_by_age_class.plot(kind='bar', figsize=(10, 6))
  plt.title('Survival Rate by Age Group and Passenger Class')
  plt.xlabel('Age Group')
  plt.ylabel('Mean Survival Rate')
  plt.legend(title='Passenger Class')
  plt.show()
```
