Problem - 1: Array Creation:
Complete the following Tasks:
1. Initialize an empty array with size 2X2.
2. Initialize an all one array with size 4X2.
3. Return a new array of given shape and type, filled with fill value.{Hint: np.full}
4. Return a new array of zeros with same shape and type as a given array.{Hint: np.zeros like}
5. Return a new array of ones with same shape and type as a given array.{Hint: np.ones like}
6. For an existing list new_list = [1,2,3,4] convert to an numpy array.{Hint: np.array()}


```python
import numpy as np

# Task 1: Initialize an empty array with size 2X2.
empty_array = np.empty((2, 2))
print("Empty array:\n", empty_array)

# Task 2: Initialize an all one array with size 4X2.
ones_array = np.ones((4, 2))
print("\nOnes array:\n", ones_array)

# Task 3: Return a new array of given shape and type, filled with fill value.
shape = (3, 3)
fill_value = 7
full_array = np.full(shape, fill_value)
print("\nFull array:\n", full_array)


# Task 4: Return a new array of zeros with same shape and type as a given array.
example_array = np.array([[1, 2], [3, 4]], dtype=np.int32)
zeros_like_array = np.zeros_like(example_array)
print("\nZeros like array:\n", zeros_like_array)

# Task 5: Return a new array of ones with same shape and type as a given array.
ones_like_array = np.ones_like(example_array)
print("\nOnes like array:\n", ones_like_array)

# Task 6: Convert a list to a NumPy array.
new_list = [1, 2, 3, 4]
numpy_array = np.array(new_list)
print("\nNumpy array from list:\n", numpy_array)
```

```python
import numpy as np

# Task 1: Create an array with values ranging from 10 to 49.
arr1 = np.arange(10, 50)
print("Array with values from 10 to 49:\n", arr1)

# Task 2: Create a 3x3 matrix with values ranging from 0 to 8.
arr2 = np.arange(9).reshape(3, 3)
print("\n3x3 matrix with values from 0 to 8:\n", arr2)

# Task 3: Create a 3x3 identity matrix.
arr3 = np.eye(3)
print("\n3x3 identity matrix:\n", arr3)

# Task 4: Create a random array of size 30 and find the mean.
arr4 = np.random.random(30)
mean_arr4 = arr4.mean()
print("\nRandom array of size 30:\n", arr4)
print("\nMean of the array:\n", mean_arr4)

# Task 5: Create a 10x10 array with random values and find min/max.
arr5 = np.random.random((10, 10))
min_arr5 = arr5.min()
max_arr5 = arr5.max()
print("\n10x10 array with random values:\n", arr5)
print("\nMinimum value:\n", min_arr5)
print("\nMaximum value:\n", max_arr5)

# Task 6: Create a zero array of size 10 and replace the 5th element with 1.
arr6 = np.zeros(10)
arr6[4] = 1  # 5th element (index 4)
print("\nZero array with 5th element replaced by 1:\n", arr6)

# Task 7: Reverse an array.
arr7 = [1, 2, 0, 0, 4, 0]
reversed_arr7 = arr7[::-1]
print("\nReversed array:\n", reversed_arr7)

# Task 8: Create a 2D array with 1 on the border and 0 inside.
arr8 = np.ones((5,5))
arr8[1:-1,1:-1] = 0
print("\n2D array with 1 on border and 0 inside:\n", arr8)

# Task 9: Create an 8x8 matrix with a checkerboard pattern.
arr9 = np.zeros((8, 8), dtype=int)
arr9[1::2, ::2] = 1
arr9[::2, 1::2] = 1
print("\n8x8 matrix with checkerboard pattern:\n", arr9)
```

Complete all the task using numpy:

```python
import numpy as np

x = np.array([[1,2],[3,5]])
y = np.array([[5,6],[7,8]])
v = np.array([9,10])
w = np.array([11,12])

# 1. Add the two arrays
print("Sum of x and y:\n",np.add(x,y))

#2. Subtract the two arrays
print("\nDifference of x and y:\n",np.subtract(x,y))

# 3. Multiply the array with an integer
print("\nx multiplied by 2:\n",x*2)
print("\ny multiplied by 5:\n",y*5)


# 4. Find the square of each element
print("\nSquare of x:\n",np.square(x))
print("\nSquare of y:\n",np.square(y))


# 5. Dot product
print("\nDot product of v and w:",np.dot(v,w))
print("\nDot product of x and v:\n",np.dot(x,v))
print("\nDot product of x and y:\n",np.dot(x,y))


# 6. Concatenate
print("\nConcatenate x and y along rows:\n",np.concatenate((x,y),axis=0)) #row wise
print("\nConcatenate v and w along columns:\n",np.concatenate((v[:,np.newaxis],w[:,np.newaxis]),axis=1)) #column wise, need to reshape v and w first

# 7. Concatenate x and v
# print("\nConcatenate x and v:\n",np.concatenate((x,v),axis=0))
# The above code will throw an error because x and v do not have the same number of dimensions.
# x is a 2D array (2 rows, 2 columns)
# v is a 1D array (2 elements)
# When concatenate along axis=0 (row wise) both array should have same number of column
# To concatenate, the arrays need to have compatible dimensions.
# One way to fix is this.
print("\nConcatenate x and v:\n",np.concatenate((x,v.reshape(1,2)),axis=0))
```

```python
import numpy as np
import time

# Python Lists Matrix Multiplication
def matrix_multiply_lists(A, B):
  rows_A = len(A)
  cols_A = len(A[0])
  rows_B = len(B)
  cols_B = len(B[0])

  if cols_A != rows_B:
    raise ValueError("Matrices cannot be multiplied")

  result = [[0 for _ in range(cols_B)] for _ in range(rows_A)]
  for i in range(rows_A):
    for j in range(cols_B):
      for k in range(cols_A):
        result[i][j] += A[i][k] * B[k][j]
  return result

# NumPy Arrays Matrix Multiplication
def matrix_multiply_numpy(A, B):
  return np.dot(A, B)

# Matrix sizes
size = 1000

# Initialize matrices using lists
matrix_A_list = [[j for j in range(size)] for i in range(size)]
matrix_B_list = [[j for j in range(size)] for i in range(size)]

# Initialize matrices using NumPy arrays
matrix_A_np = np.random.rand(size, size)
matrix_B_np = np.random.rand(size, size)


# Python Lists
start_time = time.time()
result_list = matrix_multiply_lists(matrix_A_list, matrix_B_list)
end_time = time.time()
print(f"Time taken for Python lists matrix multiplication: {end_time - start_time} seconds")

# NumPy Arrays
start_time = time.time()
result_np = matrix_multiply_numpy(matrix_A_np, matrix_B_np)
end_time = time.time()
print(f"Time taken for NumPy arrays matrix multiplication: {end_time - start_time} seconds")
```
