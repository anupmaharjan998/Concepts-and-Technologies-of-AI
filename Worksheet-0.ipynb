```python
temperatures = [3, 7, 13, 18, 19, 21, 4, 12, 15, 18, 9, 20, 16, 6, 14, 22, 11, 10, 7, 5, 13, 15, 19, 18]

#Task-1
cold = []
mild = []
comfortable = []

for temp in temperatures:
    if temp < 10:
        cold.append(temp)
    elif 10 <= temp <= 15:
        mild.append(temp)
    elif 15 < temp <= 20:
        comfortable.append(temp)

print("Cold temperatures:", cold)
print("Mild temperatures:", mild)
print("Comfortable temperatures:", comfortable)

#Task-2
print("Number of times it was mild:", len(mild))
print("Number of times it was comfortable:", len(comfortable))
print("Number of times it was cold:", len(cold))

#Task-3
temperatures_fahrenheit = [(temp * 9/5) + 32 for temp in temperatures]
print("Temperatures in Fahrenheit:", temperatures_fahrenheit)

#Task-4
import matplotlib.pyplot as plt
night_temps = []
evening_temps = []
day_temps = []

for i, temp in enumerate(temperatures):
    if 0 <= i % 24 < 8:
        night_temps.append(temp)
    elif 8 <= i % 24 < 16:
        evening_temps.append(temp)
    elif 16 <= i % 24 < 24:
        day_temps.append(temp)

if day_temps:
    avg_day_temp = sum(day_temps) / len(day_temps)
    print("Average day-time temperature:", avg_day_temp)

plt.plot(range(len(day_temps)), day_temps, label="Day temperatures")
plt.xlabel("Time of Day")
plt.ylabel("Temperature (°C)")
plt.title("Daytime Temperature Patterns")
plt.legend()
plt.show()
```

```python
def sum_nested_list(nested_list):
    total = 0
    for element in nested_list:
        if isinstance(element, list):
            total += sum_nested_list(element)
        else:
            total += element
    return total

nested_list = [1, [2, [3, 4], 5], 6, [7, 8]]
print("Sum of nested list:", sum_nested_list(nested_list))
```

```python
def generate_permutations(s):
    if len(s) <= 1:
        return [s]
    
    permutations = []
    for i, char in enumerate(s):
        if char not in s[:i]:
            for perm in generate_permutations(s[:i] + s[i+1:]):
                permutations.append(char + perm)
    return permutations

print("Permutations of 'abc':", generate_permutations("abc"))
print("Permutations of 'aab':", generate_permutations("aab"))
```

```python
def calculate_directory_size(directory):
    total_size = 0
    for key, value in directory.items():
        if isinstance(value, dict):
            total_size += calculate_directory_size(value)
        else:
            total_size += value
    return total_size

directory_structure = {
    "file1.txt": 200,
    "file2.txt": 300,
    "subdir1": {
        "file3.txt": 400,
        "file4.txt": 100
    },
    "subdir2": {
        "subsubdir1": {
            "file5.txt": 250
        },
        "file6.txt": 150
    }
}
print("Total directory size:", calculate_directory_size(directory_structure))
```

```python
def min_coins(coins, amount):
    dp = [float('inf')] * (amount + 1)
    dp[0] = 0

    for coin in coins:
        for i in range(coin, amount + 1):
            dp[i] = min(dp[i], dp[i - coin] + 1)

    return dp[amount] if dp[amount] != float('inf') else -1

coins = [1, 2, 5]
amount = 11
print("Minimum coins to make amount 11:", min_coins(coins, amount))

coins = [2]
amount = 3
print("Minimum coins to make amount 3:", min_coins(coins, amount))
```

```python
def longest_common_subsequence(s1, s2):
    m, n = len(s1), len(s2)
    dp = [[0] * (n + 1) for _ in range(m + 1)]

    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if s1[i - 1] == s2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1] + 1
            else:
                dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])

    return dp[m][n]

s1 = "abcde"
s2 = "ace"
print("Length of LCS ('abcde', 'ace'):", longest_common_subsequence(s1, s2))

s1 = "abc"
s2 = "ab"
print("Length of LCS ('abc', 'abc'):", longest_common_subsequence(s1, s2))

s1 = "abce"
s2 = "def"
print("Length of LCS ('abce', 'def'):", longest_common_subsequence(s1, s2))
```
